## dimension reduction

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(n_components=2)
%time proj = pca.fit_transform(xx_train.todense())
```

```python
plt.scatter(proj[:, 0], proj[:, 1], c=yy_train)
plt.colorbar()
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2)
proj = svd.fit_transform(xx_train) 
```

```python
plt.scatter(proj[:, 0], proj[:, 1], c=yy_train)
plt.colorbar()
```

## Tsne

first use truncatedSVD to reduce dimension to 50 then apply TSNE

```python
from sklearn.decomposition import TruncatedSVD

X_reduced = TruncatedSVD(n_components=50, random_state=0).fit_transform(xx_train)
```

```python
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2, perplexity=40, verbose=2).fit_transform(X_reduced)
```
